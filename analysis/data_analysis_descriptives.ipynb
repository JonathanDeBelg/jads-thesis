```python
import pandas as pd

igp = 'scraper/player-stats/fbref.com/data/def_match_data_preprocessed.csv'

igp_df = pd.read_csv(igp)
igp_df
```

```python
metrics = ['Gls', 'Ast', 'PK', 'PKatt', 'Sh', 'SoT', 'CrdY', 'CrdR', 'Touches', 'Tkl', 'Int', 'Blocks']
igp_df = igp_df[['Player Name', ''] + metrics]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

igp_df = igp_df.fillna(0)

scaler = StandardScaler()
data_standardized = pd.DataFrame(scaler.fit_transform(igp_df[metrics]), columns=metrics)

# Perform PCA
pca = PCA(n_components=1)
igp_df['Composite_Score'] = pca.fit_transform(data_standardized[metrics])
```
