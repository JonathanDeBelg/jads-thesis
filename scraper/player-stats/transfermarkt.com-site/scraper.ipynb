```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Specify the URL
def scrape_data(urls):
    all_matches = []

    for url in urls:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        boxes = soup.find_all("div", class_="box")
        player_name = soup.find("h1").text.strip()


        for box in boxes[1:]:
            competition = box.find("div", class_="content-box-headline")
            if competition:
                competition = competition.text.strip()
                table = box.find("table")
                if table:
                    headers = ['Matchday', 'Date', 'Home team country', 'Home team', 'Away team country', 'Away team', 'Result', 'Pos.', 'Goals', 'Assists', 'Own goals', 'Yellow cards', 'Second yellow cards', 'Red cards', 'Substitutions on', 'Substitutions off', 'Minutes played', 'Competition', 'Player name']

                    for row in table.find("tbody").find_all("tr"):
                        cols = row.find_all("td", recursive=False)
                        match_data = [col.text.strip() for col in cols]

                        if any(c in row.get("class") for c in ("bg_rot_20", "bg_gelb_20")):
                            needed_fields = len(headers) - 2 - len(match_data)
                            if needed_fields > 0:
                                for amount in range(needed_fields):
                                    match_data.append('')

                        match_data.append(competition)
                        match_data.append(player_name)
                        all_matches.append(dict(zip(headers, match_data)))

    df = pd.DataFrame(all_matches)
    return df
```

```python
# Example usage with multiple URLs
urls = [
    'https://www.transfermarkt.co.uk/lutsharel-geertruida/leistungsdaten/spieler/420210/saison/2023/plus/1',
    'https://www.transfermarkt.co.uk/justin-bijlow/leistungsdaten/spieler/315121/saison/2023/plus/1',
    'https://www.transfermarkt.co.uk/timon-wellenreuther/leistungsdatendetails/spieler/196829/saison/2023/plus/1'
]

df = scrape_data(urls)
```
